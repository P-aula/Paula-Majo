```python
def preprocess(text):
    # Aquí puedes agregar más optimizaciones si es necesario
    tokens = word_tokenize(text.lower())
    tokens = [word for word in tokens if word.isalnum()]  # Filtrar solo palabras
    return ' '.join(tokens)
```

# Modelo de Embeddings con FastText y K-Means

```python
import fasttext
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
import fasttext

# Concatenar texto
input_text = data['processed_text'].str.cat(sep='\n')

# Guardar en un archivo temporal si hay problemas con cadenas largas
with open("temp_text.txt", "w", encoding="utf-8") as f:
    f.write(input_text)

# Entrenar el modelo
model = fasttext.train_unsupervised("temp_text.txt", model='skipgram', epoch=5)

# Guardar el modelo
model.save_model("model.bin")
```

```python
data['embedding'] = data['processed_text'].apply(lambda x: model.get_sentence_vector(x))

```

```python
pca = PCA(n_components=8)
reduced_embeddings = pca.fit_transform(np.vstack(data['embedding'].values))
```

```python
from sklearn.cluster import KMeans
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Aplicar KMeans
kmeans = KMeans(n_clusters=5, init='k-means++', n_init=10, max_iter=300, random_state=42)
data['cluster'] = kmeans.fit_predict(reduced_embeddings)

# Asegúrate de que 'data_text' tenga los datos actualizados
data_text['cluster'] = data['cluster']

# Generar nubes de palabras por clúster
for cluster in range(5):  # Cambia el rango si tienes más clústeres
    cluster_texts = ' '.join(
        [' '.join(text) for text in data[data['cluster'] == cluster]['processed_text']]
    )
    wordcloud = WordCloud(width=800, height=400).generate(cluster_texts)
    plt.figure()
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud for the Cluster {cluster}')
    plt.show()
```

```python
def assign_label(cluster_num):
    labels = {0: 'Politics', 1: 'Economy', 2: 'Social'}  # Ejemplo de etiquetas
    return labels.get(cluster_num, 'Other')

data_text['category'] = data_text['cluster'].apply(assign_label)
print(data_text[['headline_text', 'category']].head())
```

```python
from gensim import corpora

# Crear el diccionario
dictionary = corpora.Dictionary(data['processed_text'])

# Crear el corpus en formato Bag-of-Words (BoW)
corpus = [dictionary.doc2bow(text) for text in data['processed_text']]

# Verificar los resultados
print(dictionary)  # Muestra palabras únicas e índices
print(corpus[:5])  # Muestra el BoW de los primeros 5 documentos
```

```python
lda_model = LdaModel(corpus, num_topics=5, id2word=dictionary, passes=5)
```

```python
for idx, topic in lda_model.print_topics(-1):
    print(f'Topic: {idx} \nWords: {topic}')
```

```python
def assign_topic_label(lda_model, corpus):
    topic_labels = {0: 'Politics', 1: 'Economy', 2: 'Culture'}
    topic_assignments = []
    for doc in corpus:
        topics = lda_model.get_document_topics(doc)
        top_topic = max(topics, key=lambda x: x[1])[0]
        topic_assignments.append(topic_labels.get(top_topic, 'Other'))
    return topic_assignments
```

```python
data_text['lda_category'] = assign_topic_label(lda_model, corpus)
print(data_text[['headline_text', 'lda_category']].head())
```

# Modelo BERT


```python
class NewsDataset(Dataset):
    def __init__(self, texts):
        self.texts = texts

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        return self.texts[idx]
```

```python
# Cargar el modelo BERT preentrenado
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
model.to('cuda')
```

```python
dataset = NewsDataset(data_text['headline_text'].tolist())
dataloader = DataLoader(dataset, batch_size=32)

embeddings = []
for batch in dataloader:
    inputs = tokenizer(batch, return_tensors='pt', padding=True, truncation=True, max_length=512)
    inputs = {key: val.to('cuda') for key, val in inputs.items()}  # Asegurarse de que los inputs estén en GPU
    outputs = model(**inputs)
    embeddings.append(outputs.last_hidden_state.mean(dim=1).detach().cpu().numpy())
```

```python
import numpy as np
import torch
from torch.utils.data import DataLoader

# Asegúrate de que la clase NewsDataset esté definida correctamente
dataset = NewsDataset(data_text['headline_text'].tolist())
dataloader = DataLoader(dataset, batch_size=32)

embeddings = []

# Procesar el DataLoader y generar los embeddings
for batch in dataloader:
    inputs = tokenizer(batch, return_tensors='pt', padding=True, truncation=True, max_length=512)
    inputs = {key: val.to('cuda') for key, val in inputs.items()}  # Asegurarse de que los inputs estén en GPU

    # Pasar los inputs al modelo para obtener los embeddings
    outputs = model(**inputs)

    # Obtener el embedding promedio por token en la capa de salida (última capa oculta)
    batch_embeddings = outputs.last_hidden_state.mean(dim=1).detach().cpu().numpy()

    # Almacenar los embeddings en la lista
    embeddings.append(batch_embeddings)

# Convertir la lista de embeddings a un arreglo numpy
embeddings_array = np.concatenate(embeddings, axis=0)

# Guardar los embeddings en un archivo .npy
np.save('embeddings.npy', embeddings_array)

print(f"Embeddings guardados en 'embeddings.npy'. Forma del arreglo: {embeddings_array.shape}")
```
